## VIF Testing Our Dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaler = scaler.fit_transform(x)
```

```python
x_scaler
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x_scaler.shape[1]
```

```python
vif = pd.DataFrame()
vif["VIF Score"] = [variance_inflation_factor(x_scaler, i) for i in range(x_scaler.shape[1])]
vif["Columns"] = x.columns
```

```python
vif
```

Hence, we have the correlation values for all the features. As a thumb rule,

If VIF value greater than 5 means a very severe multicollinearity

Here we have tax and rad feature whose vif score is greater than 5, hence we drop them.

```python

x_new = pd.DataFrame(x_scaler, columns = x.columns)
```

## Lets split our dataset into train & test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size = .25, random_state= 234)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
regression.score(x_train, y_train)
```

```python
regression.score(x_test, y_test)
```

```python
y_pred = regression.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

## Let's see if our Model is overfitting our traing data

### Lasso Regularization

LassoCv will return best alpha and coefficent after performing 10 cross validation

```python
from sklearn.linear_model import LassoCV, Lasso,RidgeCV, Ridge, ElasticNetCV, ElasticNet
```

```python
lassocv = LassoCV(alphas = None, cv = 10, max_iter = 100000, normalize = True)
lassocv.fit(x_train, y_train)
```

```python
# best alpha parameter
alpha = lassocv.alpha_
alpha
```

```python
#now that we have best parameter, let's use Lasso regression and see how well our data has fitted before
lasso_reg=Lasso(alpha)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

our r2_score for test data (75.34%) comes same as before using regularization. So, it is fair to say our OLS model did not overfit the data.


### Ridge Regularization

```python
# RidgeCV will return best alpha and coefficients after performing 10 cross validations. 
# We will pass an array of random numbers for ridgeCV to select best alpha from them
alphas=np.random.uniform(low=0,high=10,size=(50,))
ridge_cv=RidgeCV(alphas=alphas,cv=10,normalize=True)
ridge_cv.fit(x_train,y_train)
```

```python
#best possible alpha value
ridge_cv.alpha_
```

```python
ridge_model=Ridge(alpha=ridge_cv.alpha_)
ridge_model.fit(x_train,y_train)
```

```python
ridge_model.score(x_test,y_test)
```

### Elastic Net

```python
elastic_cv=ElasticNetCV(alphas=None,cv=10)
elastic_cv.fit(x_train,y_train)
```

```python
elastic_cv.alpha_
```

```python
# l1_ration gives how close the model is to L1 regularization, below value indicates we are giving equal preference to L1 and L2
elastic_cv.l1_ratio
```

```python
elastic_reg=ElasticNet(alpha=elastic_cv.alpha_,l1_ratio=0.5)
elastic_reg.fit(x_train,y_train)
```

```python
elastic_reg.score(x_test,y_test)
```

So, we can see by using different type of regularization, we still are getting the same r2 score. That means our OLS model has been well trained over the training data and there is no overfitting.
